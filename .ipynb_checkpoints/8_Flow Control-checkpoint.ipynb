## 1.3) if-elif-else

```python
brand  = input("Enter your favourite brand: ")

if brand == "rc":
    print("it is childrens brand")
elif brand == 'kf':
    print("it is not that much kick")
elif brand == "ko":
    print("it is too light")
elif brand == "fo":
    print("buy one get one free")
else:
    print("other brands are not recommended")
```

# 2. Iterative Statements : (for loop , while loop)

## 2.1) for loop

```python
name = "vinod shende"
count = 0
for x in name:
    count += 1
    print(x)
    
print("\nnumber of characters: ",count)
```

```python
l = [10,20,20,40,50]

for x in l:
    print(x)
```

```python
for x in range(1,11):
    print(x)
```

```python
name = input("enter the name : ")
index = int(input("enter the index : "))

if index <= len(name)-1:
    for x in range(0,len(name)):
        if index == x:
            print("the character present at index",index,"is",name[index])

else:
    print("index out of range")
```

```python
s = input("enter string : ")
i = 0

for x in s:
    print("the character present at index",i,"is",x)
    i += 1 
```

```python
for x in range(21):
    if x%2 != 0:
        print(x)
```

```python
l = eval(input("enter the list: "))
sum = 0
for x in l:
    sum += x
    
print("total addition : ",sum)
```

## 2.2) while loop

```python
x = 1

while x <= 10:
    print(x)
    x += 1
```

```python
n = int(input("Enter the number: "))
sum = 0
i = 1
while i <= n:
    sum += i
    i += 1
print(sum)
    
```

```python
name = input("enter the name: ")

while name != "vinod":
    print("Enter right name")
    name = input("enter the name: ")
print("thanks for confirmation")
```

```python
username = input("Enter the username : ")
password = input("Enter the password : ")

while username != "Vinod" and password != 12345:
    
    print("\nIncorrect username or password!!!!!\nEnter the correct details\n")
    
    username = input("Enter the username : ")
    password = input("Enter the password : ")
    
print("Thanks for registration")
```

## Nested Loops :

```python
for i in range(4):
    for j in range(4):
        print(f"i = {i} and j = {j}")
```

```python
n = int(input("Number of  rows: "))

for i in range(1,n+1):         # i represents row_number
    for j in range(1,i+1):     # j represents no. of * in row
        print("*", end = " ")
    print()
```

```python
n = int(input("Number of  rows: "))

for i in range(1,n+1):         # i represents row_number
    print("* "*i, end = " ")
    print()
```

```python
n = int(input("Number of  rows: "))

for i in range(1,n+1): # i represents row_number
        print("* "*n, end = " ")
        print()
```

```python
n = int(input("Number of  rows: "))

for i in range(n):         # i represents row_number
    for j in range(n):     # j represents no. of * in row
        print("*", end = " ")
    print()
```

# 3. Transfer Statements (break, continue, pass)

## 3.1) break

```python
for i in range(10):
    if  i == 7:
        print("processing is enough.....please break!!!!!!!!!!")
        break
    print(i)
```

```python
l = [12,34,456,10,57,92,500,2788,6000]

for x in l:
    if x > 500:
        print(f"Rs.{x} too expensive.....!!!!!")
        break
    print(f"item of Rs.{x} bought!!")
```

## 3.2) continue

```python
  for i in range(10):
        if i % 2 == 0:
            continue
        print(i)
```

```python
l = [12,34,456,10,57,92,500,2788,6000,34,56]

for x in l:
    if x > 500:
        print(f"Rs.{x} too expensive.....!!!!!")
        continue
    print(f"item of Rs.{x} bought!!")
```

```python
num = [10,0,20,50,0,40,0]

for n in num:
    if n == 0:
        print("\nZeroDivisionError!!!!!!!!!!!\n")
        continue
    print(f"100/{n} = {100/n}")
```

## Loops with else block :

#### Inside loop execution (for / while / try-except-finally), if break statement not executed, then only else part will be executed. 


#### 1. for - else          2. while - else         3. try - except - else - finally

```python
l = [12,34,456,10,57,92,500,2788,6000]

for x in l:
    if x > 500:
        print(f"Rs.{x} too expensive.....!!!!!")
        break
    print(f"item of Rs.{x} bought!!")
else:
    print("Congrats....All items are processed")
```

```python
l = [12,34,456,10,57,92,500]

for x in l:
    if x > 500:
        print(f"Rs.{x} too expensive.....!!!!!")
        break
    print(f"item of Rs.{x} bought!!")
else:
    print("Congrats....All items are processed")
```
